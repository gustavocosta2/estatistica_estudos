# **Covariância e Correlações**

## **1. Covariância**

A covariância é uma medida do grau de interdependência numérica entre duas variáveis. Ela indica a direção da relação linear entre as variáveis. Se for **positiva**, eles tendem a se mover na mesma direção. Se for **negativa**, uma variável tende a aumentar enquanto a outra diminui.

formula-covariancia.webp

### **Características**

- Mede a relação linear entre duas variáveis. Valores positivos ou negativos indicam a **direção** dessa relação, mas não a força.

**USO:** Útil para determinar se uma relação linear existe, mas não é ideal para comparações diretas devido à dependência de unidades de medida.

**Exemplo:** Se a temperatura aumenta e as vendas de picolé aumentam também, a covariância será positiva. Mas há uma observação, isso não é uma relação de causalidade! A temperatura aumentar não causa o aumento das vendas de picolé!

### **Covariância em Código**

````python
# Calculando a covariância em Python sem o uso de bibliotecas.

# Função calculadora da média de uma lista.
def media(lista):
  return sum(lista) / len(lista)

# Função para calcular a covariância.
def covariance(lista1, lista2, sample=False):

  if(sample):
    denominador = len(lista1) - 1
  else:
    denominador = len(lista1)

  media_x = media(lista1)
  media_y = media(lista2)

  soma = 0
  for elemento_x, elemento_y  in zip(lista1, lista2):
    soma += (elemento_x - media_x) * (elemento_y - media_y)

  return soma / denominador
````

````python
# Dados para calcular a covariância.
idade = [25, 30, 35, 40, 45, 50, 55, 60, 65]
salario = [40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000]

# Cálculo da covariância.
covariancia = covariance(idade, salario)

print(f'Covariância de idade e salário: {covariancia}')
````

## **2. Correlações**

A correlação é uma medida estatística que descreve a extensão da relação entre duas variáveis.

Aqui, serão abordadas três correlações:

- Correlação de Pearson;
- Correlação de Kendall;
- Correlação de Spearman.


### **2.1. Correlação de Pearson**

A correlação de Pearson mede a relação linear entre duas variáveis **contínuas**, variando de -1 até +1. Sendo que:

- **-1** indica que há uma correlação negativa perfeita.
- **+1** indica que há uma correlação positiva perfeita.

**Exemplo:** Suponha duas variáveis X (idade) e Y (pressão arterial). Se idades mais altas possuem pressões arteriais elevadas e a correlação calculada é 0.6, isso indica uma correlação positiva moderada entre idade e pressão arterial.

#### **Correlação de Pearson em Código**

````python
# Função calculadora da média.
def media(lista):
  return sum(lista)/len(lista)

# Funcão calculadora da variância.
def variance(lista):
  media = media(lista)
  soma = 0
  for elemento in lista:
    soma += (elemento - media)**2

  return soma ** 0.5
````

````python
# Função calculadora da Correlação de Pearson.

def corr_pearson(listaX, listaY):
  soma = 0
  media_x = media(listaX)
  media_y = media(listaY)

  for elemento_x, elemento_y in zip(listaX, listaY):
    soma += (elemento_x - media_x) * (elemento_y - media_y)

  return soma / (variance(listaX) * variance(listaY))
````

### **2.2. Correlação de Kendall**

Mede a força e a direção da associação entre duas variáveis, baseada na concordância entre os pares de observações.

**USO:** Útil para dados ordinais ou pequenos conjuntos de dados.


**Características:** Avalia a associação baseada na concordância entre os pares de observações, sendo **menos sensível a outliers** do que a correlação de Pearson. Também varia de -1 a +1.

correlacao-kendall.jpg

**Exemplo:** Se classificarmos 10 filmes por duas pessoas e encontrarmos que em 45 pares possíveis de filmes:

- 30 são concordantes (significa que as duas pessoas classificaram o mesmo filme mais alto que o outro filme ou mais baixo que o outro filme)

- 15 são discordantes.

Logo o Tau de Kendall seria calculado

```
(30 - 15) / 45 = 0.33
```

Logo há uma correlação positiva fraca.

#### **Correlação de Kendall em Código**

````python
# Importação da biblioteca scipy com a função calculadora da correlação de Kendall.
from scipy.stats import kendalltau

kendalltau(idade, salario)[0]
````

### **2.3. Correlação de Spearman**

Avalia a relação mónotona entre duas variáveis utilizando **ranks** ao invés de valores reais.

**USO:** É especialmente eficaz para dados não-lineares ou ordinais. Ela é robusta contra distorções nas escalas reais das medidas.

correlacao-spearman.png

Note que esse 'D' na fórmula representa a diferença dos valores nos rankings das duas variáveis Xi e Yi.

#### **Correlação de Spearman em Código**

````python
# Importação da função que calcula a correlação de Spearman.
from scipy.stats import spearmanr

spearmanr(idade, salario)[0]
````
